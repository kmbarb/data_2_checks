```python
import pandas as pd
import numpy as np
```

I am importing the Public Library Income Per Capita csv from the California State Library's webpage, and displaying the first five columns.

```python
df = pd.read_csv('https://www.library.ca.gov/wp-content/uploads/2022/11/LIPC-FY2021.csv?2022-10-12')
```

```python
print(df.shape)
df.head()
```

The column names are in the first row rather than the header, and they could be renamed for easier use in python. First I will rename the columns.

```python
fixed_columns = {
    'Unnamed: 0' : 'location',
    'Unnamed: 1' : 'population',
    'Unnamed: 2' : 'local_government_income',
    'Unnamed: 3' : 'local_income_per_capita'
}

df.rename(columns=fixed_columns, inplace=True)
df.columns
```

```python
print(df.shape)
df.head()
```

The old names were still in the first row of the dataset, so I will use iloc to remove the first row.

```python
df = df.iloc[1:]
print(df)
```
